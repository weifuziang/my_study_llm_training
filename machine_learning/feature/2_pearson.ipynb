```python
import pandas as pd
advertising = pd.read_csv(r"F:\MyStudy\LLM\my_codes\data\advertising.csv")
print(advertising)
#axis=1 按行每一行删除； inplace=True在原有的基础上删除，不做新参数的接受；
advertising.drop(advertising.columns[0],axis=1,inplace=True)
#删除空数据
advertising.dropna(inplace=True)
print(advertising)
X = advertising.drop("Sales", axis=1)
y = advertising["Sales"]
```

```python
#计算皮尔逊相关系数：X是特征向量，y是标签，相关数值越大就越相关
print(X.corrwith(y,method="pearson"))
```

```python
#皮尔逊相关系数计算（不区分X和y的）: 会出现自己和自己的相关系数值，即为1；
corr_matrix = advertising.corr(method="pearson")
print(corr_matrix)
```

```python
#将相关系数矩阵画成热力图
import matplotlib.pyplot as plt
import seaborn as sns
#annot=True:开启数值；fmt=".2f"：精确两位； cmap="coolwarm"：冷热图显示；
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```
