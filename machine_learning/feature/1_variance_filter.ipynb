```python
import numpy as np
```

```python
#===========================================低方差过滤法=======================================
#构建特征

#特征1
#randn 是默认服从 均值u=0  方差为1的正态分布
a = np.random.randn(100)
#打印方差
print(np.var(a))
```

```python
#特征2
#改变了方差：0.1的平方，即0.01
#np.random.randn(100)*0.1
#均值为u=5,方差为0.01 ，每次执行的结果都不太一样
b = np.random.normal(5,0.1,size=100)
#打印方差
print(np.var(b))
```

```python
# 构造特征向量（输入数据X）

#数据方向的堆叠
X = np.vstack((a,b)).T
print(X)
print(X.shape)
```

```python
# 低方差过滤
from sklearn.feature_selection import VarianceThreshold

#方差大于0.01的向量
vt = VarianceThreshold(0.01)
#受上面a 和 b 执行结果的影响
vt_filtered = vt.fit_transform(X)
print(vt_filtered)
print(vt_filtered.shape)
```
