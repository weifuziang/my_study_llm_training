```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 生成 y = sinx
x = np.arange(0,6,0.1)
y = np.sin(x)
print(x,y)
```

```python
# 画图
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = sinx')
plt.show()
```

```python
# 画出 y' = cosx
y1 = np.cos(x)
plt.plot(x,y1)
plt.xlabel('x')
plt.ylabel('y')
plt.title("y' = cosx")
plt.show()
```

```python
plt.plot(x,y,label='y=sinx')
plt.plot(x,y1,linestyle="--",label='y=cosx')
plt.xlabel('x')
plt.ylabel('y')
plt.title("y' = sinx")
plt.show()
```
