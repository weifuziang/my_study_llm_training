```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 生成 y = sinx
x = np.arange(0,6,0.1)
y = np.sin(x)
print(x,y)
```

```python
# 画图
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = sinx')
plt.show()
```

```python
# 画出 y' = cosx
y1 = np.cos(x)
plt.plot(x,y1)
plt.xlabel('x')
plt.ylabel('y')
plt.title("y' = cosx")
plt.show()
```

```python
plt.plot(x,y,label='y=sinx')
plt.plot(x,y1,linestyle="--",label='y=cosx')
plt.xlabel('x')
plt.ylabel('y')
plt.title("y' = sinx")
plt.show()
```

```python
# ==================================================================向量的计算=====================================================
import numpy as np

x = np.array([2, 5, 8])
#向量打印
print(x)
#向量的转置
print(x.T)
#向量的形状
print(x.shape)
```

```python
y = np.array([1, 3, 7])
#向量加法
print(x+y)
print(x*3)

#向量乘法
print(x * y)
```

```python
# 向量的点积 或者内积
print(x.dot(y))
print(x @  y)
```

```python
# 求向量的范数 l0范数 ： 非零元素的个数
print(x)
l0_norm=np.linalg.norm(x,ord=0)
print(l0_norm)
```

```python
# # 求向量的范数 l1范数： 直角线的距离
l1_norm = np.linalg.norm(x,ord=1)
print(l1_norm)
```

```python
# # 求向量的范数 l1范数： 欧几里得范数，即直角坐标的斜边长度
x = np.array([2, 5, 8])

l2_norm = np.linalg.norm(x,ord=2)
print(l2_norm)
```

```python
#矩阵的初始化
import numpy as np
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[1,3,7],[5,0,2]])
print("A.shape: ",A.shape)
print(A)
print(A.T)
print("B.shape: ",B.shape)
print(B)
print(B.T)
```

```python
# 矩阵加法: 对应位置相加
print(A+B)
```

```python
#矩阵转置 ：（A+B).T = A.T + B.T
print((A+B).T)
print(A.T + B.T)
```

```python
print(A)
print(3*A)
```

```python
#矩阵的哈达玛积（Hadamard）： 对应位置相乘
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[1,3,7],[5,0,2]])
print(A)
print(B)
print("A * B: \n", A * B)
print("B * A: \n", B * A)

# 矩阵的内积：对应位置相乘之和，得到的是一个标量
print("B和A的内积", np.sum(B * A))
```

```python
# ***矩阵的乘法：行列相乘
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[1,3,7],[5,0,2]])
print(A.dot(B.T))
print(A @ B.T)

# 矩阵的向量化: 把矩阵的每一行都追加到第一列上
print(A.flatten())
```

```python
#矩阵的克罗内克积（Kronecker）: A矩阵的每个元素，和B矩阵的相乘
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[1,3,7],[5,0,2]])
print(np.kron(A,B))
```

```python
#矩阵的逆： 被逆的矩阵必须是方阵
C = np.array([[1,2],[3,5]])
c_inv = np.linalg.inv(C)
c_inv = np.around(c_inv).astype(int)
print(np.linalg.inv(C))
print(c_inv @ C)
```

```python
#伪逆矩阵：A本身不是一个方阵，那么怎么求矩阵的逆矩阵呢
A = np.array([[1,2,3],[4,5,6]])

A_inv = np.linalg.pinv(A)
#A矩阵是2*3 那么他的逆矩阵就是3*2
print(A_inv)
#一定是A @ A_inv 不能是 A_inv @ A 因为 矩阵乘法满足结合律、左分配律和右分配律。不满足交换律即AB≠BA
print(A @ A_inv)
```

```python
# ==============================================矩阵求导之梯度矩阵======================================
import numpy as np

#梯度向量（实际上是一维数组，其变元只有一个行索引x，但是值一个向量） :
#函数表现形式：
#   实标量函数f是3,7,9,8 ；
#   实向量变量x是3,7,9,8各自对应的索引；
#   函数表达形式为： f(1)=3 f(2)=7 f(3)=9 f(4)=8
#利用微积分的原理计算如下
#   3位置的梯度是：（7-3）/ 1 = 4
#   7位置的梯度是：（9-3）/ 2 = 3
#   9位置的梯度是：（8-7）/ 2 = 0.5
#   8位置的梯度是： （8-9）/ 1 = -1

f = np.array([3, 7, 9, 8])
print(f)
grad = np.gradient(f)
print(grad)



```

```python
#梯度矩阵(实际上是二维数组，其变元是两个，即行索引x, 列索引y):
#函数表现形式：
#   实标量函数f是 1,3,7 5,0,2 ；
#   实矩阵变元X是1,3,7 5,0,2对应的行索引和列索引；
#   函数表达形式是f(1,1)=1 f(1,2)=3 f(1,3)=7 f(2,1)=5 f(2,2)=0 f(2,3)=2

#利用微积分原理计算矩阵梯度如下
#对列索引x1求导
#   1位置在列索引的梯度是：（5-1）/ 1 = 4
#   3位置在列索引的梯度是：（0-3）/ 1 = -3
#   7位置在列索引的梯度是：（2-7）/ 1 = -5
#   5位置在列索引的梯度是：（5-1）/ 1 = 4
#   0位置在列索引的梯度是：（0-3）/ 1 = -3
#   2位置在列索引的梯度是：（2-7）/ 1 = -5

#对行索引x2求导
#   1位置在行索引的梯度是：（3-1）/ 1 = 2
#   3位置在行索引的梯度是：（7-1）/ 2 = 3
#   7位置在行索引的梯度是：（7-3）/ 1 = 4
#   5位置在行索引的梯度是：（0-5）/ 1 = -5
#   0位置在行索引的梯度是：（5-2）/ 2 = 1.5
#   2位置在行索引的梯度是：（2-0）/ 1 = 2

B = np.array([[1,3,7],[5,0,2]])
print(B)
print(np.gradient(B))



```

```python
#梯度三维张量(实际上是三维数组，其变元是三个，即行外行对应的索引x, 内行列索引y，内列索引z)
tensor = np.array([[[5,1,8],[7,2,10],[1,5,7]],[[20,9,32],[16,9,15],[3,1,8]]])
print(tensor)
print(np.gradient(tensor))
```
